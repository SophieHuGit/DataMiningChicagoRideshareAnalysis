### Select distinct combinations of pickup and dropoff locations

```python
#First create a region with all pickup and dropoff locations

regions = project_data.loc[:,['region_pickup','long_centroid_pickup','lat_centroid_pickup']]
regions.head()
```

```python
# Change column names
regions=regions.rename(columns={'region_pickup': 'region', 'long_centroid_pickup': 'latitude',"lat_centroid_pickup":"longtitude"})
                        
```

```python
regions_dropoff=project_data.loc[:,['region_dropoff','long_centroid_dropoff','lat_centroid_dropoff']]
```

```python
regions_dropoff=regions_dropoff.rename(columns={'region_dropoff': 'region', 'long_centroid_dropoff': 'latitude',"lat_centroid_dropoff":"longtitude"})
```

```python
region_data=regions.append(regions_dropoff)
```

```python
region_data.head()
```

```python
# Now let us select distinct rows
region_data=region_data.drop_duplicates()
```

```python
#Import dataset
region_data.to_csv("region_data.csv")
```

```python
cleaned_data=pd.read_csv("cleaned_data.csv")
cleaned_data.head()
```
