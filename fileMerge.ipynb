```python
import pandas as pd
```

### Check/import data

1. Check final trip data

```python
final_trips=pd.DataFrame(pd.read_csv("WIP_data_merge_files_final_trips.csv"))
final_trips.head()
```

2. Check date data

```python
dim_date=pd.DataFrame(pd.read_csv("WIP_data_merge_files_dim_date.csv"))
dim_date.head()
```

3. Check trip lookup data

```python
trip_lookup=pd.DataFrame(pd.read_csv("WIP_data_merge_files_trip_region_lookup.csv"))
trip_lookup.head()
```

4. Check traffic data

```python
traffic=pd.DataFrame(pd.read_csv("WIP_data_merge_files_traffic_data.csv"))
traffic.head()
```

5. Check clean region data

```python
clean_region=pd.DataFrame(pd.read_csv("WIP_data_merge_files_Copy of clean_regions.csv"))
clean_region.head()
```

### Now let us join the trip data with the time data

```python
#Merge first using start_timestamp_id
merged_result=pd.DataFrame()
merged_result=pd.merge(final_trips,dim_date, left_on="start_timestamp_id",right_on="timestamp_id", how="left").drop(columns="timestamp_id")

merged_result.rename(columns={'time': 'start_time', 'date': 'start_date', "month":"start_month", "year":"start_year","weekday":"start_weekday"}, inplace=True)



```

```python
#Now merge again using end_timestamp_id
merged_result=pd.merge(merged_result,dim_date, left_on="end_timestamp_id",right_on="timestamp_id", how="left").drop(columns="timestamp_id")
merged_result.rename(columns={'time': 'end_time', 'date': 'end_date', "month":"end_month", "year":"end_year","weekday":"end_weekday"}, inplace=True)

```

```python
#Making sure the file is merged properly
merged_result.head()
```

### Now let us join the result with trip_lookup

```python
#Merge using pickup_centroid_location
merged_result2=pd.merge(merged_result,trip_lookup, left_on="pickup_centroid_location",right_on="trip_centroids", how="left")
```

```python
merged_result2.rename(columns={"trip_region_id":"pickup_trip_region_id", "base_region_id":"pickup_base_region_id"}, inplace=True)
merged_result2=merged_result2.iloc[:,0:27] #drop last column which is trip_centroids
merged_result2.head()
```

```python
#Merge using dropoff_centroid_location
```

```python
merged_result2=pd.merge(merged_result2,trip_lookup, left_on="dropoff_centroid_location",right_on="trip_centroids", how="left")

merged_result2.rename(columns={"trip_region_id":"dropoff_trip_region_id", "base_region_id":"dropoff_base_region_id"}, inplace=True)

```

```python
merged_result2=merged_result2.iloc[:,0:29] #drop centroids_location
merged_result2.head()
```

### Now merge with traffic

```python
#Merge using start_timestamp_id and pickup_base_region_id
merged_result3=merged_result2.merge(traffic, left_on=['start_timestamp_id', 'pickup_base_region_id'],right_on=['timestamp_id', 'region_id'], how="left")

merged_result3.rename(columns={"speed":"speed_pickup_start", "speed_category":"speed_category_pickup_start","bus_count":"bus_count_pickup_start", "gps_pings":"gps_pings_pickup_start"}, inplace=True)

import numpy as np
merged_result3=merged_result3.iloc[:,np.r_[0:29,32:36]] #delete unwanted columns
```

```python
merged_result3.head()
```

```python
#Merge using start_timestamp_id and dropoff_base_region_id
merged_result4=merged_result3.merge(traffic, left_on=['start_timestamp_id', 'dropoff_base_region_id'],right_on=['timestamp_id', 'region_id'], how="left")

merged_result4.rename(columns={"speed":"speed_dropoff_start", "speed_category":"speed_category_dropoff_start","bus_count":"bus_count_dropoff_start", "gps_pings":"gps_pings_dropoff_start"}, inplace=True)

merged_result4=merged_result4.iloc[:,np.r_[0:33,36:40]]  #delete unwanted columns 
merged_result4.head()
```

```python
#Merge using end_timestamp_id and pickup_base_region_id  
merged_result5=merged_result4.merge(traffic, left_on=['end_timestamp_id', 'pickup_base_region_id'],right_on=['timestamp_id', 'region_id'], how="left")

merged_result5.rename(columns={"speed":"speed_pickup_end", "speed_category":"speed_category_pickup_end","bus_count":"bus_count_pickup_end", "gps_pings":"gps_pings_pickup_end"}, inplace=True)

merged_result5=merged_result5.iloc[:,np.r_[0:37,40:44]] #delete unwanted columns
```

```python
merged_result5.head()
```

```python
#Merge using end_timestamp_id and dropoff_base_region_id
merged_result6=merged_result5.merge(traffic, left_on=['end_timestamp_id', 'dropoff_base_region_id'],right_on=['timestamp_id', 'region_id'], how="left")

merged_result6.rename(columns={"speed":"speed_dropoff_end", "speed_category":"speed_category_dropoff_end","bus_count":"bus_count_dropoff_end", "gps_pings":"gps_pings_dropoff_end"}, inplace=True)

merged_result6=merged_result6.iloc[:,np.r_[0:41,44:48]] #delete unwanted columns
```

```python
merged_result6.head()
```

### Now merge with clean region data

```python
#join with pickup_base_region_id
merged_result7=merged_result6.merge(clean_region, left_on="pickup_base_region_id", right_on="region_id", how="left")
merged_result7.rename(columns={"region":"region_pickup","lat_centroid":"lat_centroid_pickup","long_centroid":"long_centroid_pickup"},inplace=True)
merged_result7=merged_result7.iloc[:,np.r_[0:45,46:49]] #delete unwanted columns
```

```python
merged_result7.head()
```

```python
#join with dropoff_base_region_id
merged_result_final=merged_result7.merge(clean_region, left_on="dropoff_base_region_id", right_on="region_id", how="left")
merged_result_final.rename(columns={"region":"region_dropoff","lat_centroid":"lat_centroid_dropoff","long_centroid":"long_centroid_dropoff"},inplace=True)

merged_result_final=merged_result_final.iloc[:,np.r_[0:48,49:52]] #delete unwanted columns
```

```python
merged_result_final.head()
```

### Export file to csv

```python
merged_result_final.to_csv("merged_file.csv")
```
