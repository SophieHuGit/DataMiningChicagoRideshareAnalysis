# Classification of Tippers vs Non-Tippers with Tree Models

## Load necessary libraries

```python
# Regular libraries for data manipulation and plotting
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# Classification Tree
from sklearn import tree
# Bagging
from sklearn.ensemble import BaggingClassifier
# Random Forest
from sklearn.ensemble import RandomForestClassifier
# Boosting
from sklearn.ensemble import AdaBoostClassifier
# Model preparation
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder
# Model metrics
from sklearn import metrics
# Visualizing Classification Tree
import pydotplus
from sklearn.externals.six import StringIO  
from IPython.display import Image
```

## Load data

```python
# Load data
data = pd.read_csv("../../cleaned_data.csv", index_col = 0)
```

```python
len(data)
```

## Preparing the data for the model

```python
data.dtypes
```

```python
# Separate categorical variables
cat_vars = ["payment_type", "start_weekday", "speed_category_pickup_start", "speed_category_dropoff_end", "region_pickup", 
           "region_dropoff", "tip_flag", "ride_type", "flag_overnight", "flag_weekend"]
```

```python
# Convert into type categorial
for var in cat_vars:
    data[var] = data[var].astype('category')
```

```python
# Possible predictors
feature_names = ["duration_seconds", "miles", "payment_type", "start_month", "start_weekday", 
                 "speed_pickup_start", "speed_dropoff_end", "ride_type", "flag_overnight", "total_no_tip", 
                 "start_hour", "flag_weekend"]
# Possible target
target_name = ["tip_flag"]
```

## Encode categorical features

Variables to encode:

* Payment Type
* Start Weekday
* Ride Type

### Encode Payment Type

```python
p = list(data["payment_type"].unique())
p
```

```python
# Create map
payment_dic = {key: value for key, value in zip(p, range(1,5))}
payment_dic
```

```python
data["payment_type_encoded"] = data["payment_type"].map(payment_dic)
```

```python
data["payment_type_encoded"].unique()
```

### Encode Start Weekday

```python
data.start_weekday
```

```python
weekday_map = {key:value for key, value in zip(["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"], range(1,9))}
```

```python
weekday_map
```

```python
data["start_weekday"]
```

```python
data["start_weekday_encoded"] = data["start_weekday"].map(weekday_map)
```

```python
data["start_weekday_encoded"].head()
```

## Encode Ride Type

```python
# Create map
ride_map = {"Taxi": 1, "Rideshare": 2}
```

```python
data["ride_type_encoded"] = data["ride_type"].map(ride_map)
```

```python
data["ride_type_encoded"].head()
```

## Select features and target

```python
# Predictors
feature_names = ["duration_seconds", "miles", "payment_type_encoded", "start_month", "start_weekday_encoded", 
                 "speed_pickup_start", "speed_dropoff_end", "ride_type_encoded", "flag_overnight", "total_no_tip", 
                 "start_hour", "flag_weekend"]
# Predicted variable
target_name = ["tip_flag"]
```

## Split into Train and Test

```python
# Separate into x and y
x = data[feature_names]
y = data[target_name]
```

```python
# Split into train and test samples
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state = 232323, train_size = 0.7)
```

## Training a Classification Tree with Grid Search, on Train sample
https://stackoverflow.com/questions/35097003/cross-validation-decision-trees-in-sklearn

```python
# Create dictionary with parameters to search and its values (in this case, only max_depth of the tree)
parameters = {'max_depth':range(1,10)}
# Search over the specified values of the parameters, for an estimator (in this case, DecisionTreeClassifier)
# n_jobs indicates how many jobs to run in parallel
clf = GridSearchCV(estimator = tree.DecisionTreeClassifier(random_state = 232323), param_grid = parameters, n_jobs=4)
```

```python
# Fit the model with cross validation to estimate best max_depth
clf.fit(X = x_train, y = y_train)
```

```python
print (clf.best_score_, clf.best_params_) 
```

```python
# Build a tree with the best parameters found
tree_model = clf.best_estimator_
```

```python
# It will have the depth found in Grid Search to be optimum
tree_model.get_depth
```

## Results  
https://www.datacamp.com/community/tutorials/decision-tree-classification-python

### Accuracy in Train

```python
# Accuracy in train set
tree_model.score(x_train, y_train)
```

### Accuracy in Test

```python
# Accuracy in test set
tree_model.score(x_test, y_test)
```

### Confusion Matrix

```python
y_hat_test = tree_model.predict(x_test)
```

```python
cm = metrics.confusion_matrix(y_hat_test, y_test)
```

```python
cm = pd.DataFrame(cm)
cm.columns = ["Predicted_0", "Predicted_1"]
cm.index = ["Actual_0", "Actual_1"]
```

```python
cm
```

### Classification Report for Test data

```python
print(metrics.classification_report(y_hat_test, y_test))
```

```python
# Precision: True positive / True positive + False positive 
# People that we correctly identified as tippers out of all the ones we said would tip
round(metrics.precision_score(y_hat_test, y_test), 2)
# Precision is very bad
```

```python
# Recall or Sensitivity: True positive / True Positive + False negative 
# People that we correctly idenified as tippers out of all that actually tipped
round(metrics.recall_score(y_hat_test, y_test), 2)
# Recall is very good
```

```python
# F1 score: harmonic mean between precision and recall
round(metrics.f1_score(y_hat_test, y_test), 2)
```

## Tree plot

```python
dot_data = StringIO()

tree.export_graphviz(tree_model, 
                out_file = dot_data,  
                filled = True,
                rounded = True,
                special_characters = True, 
                feature_names = feature_names,
                class_names = ['0','1'])

graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
graph.write_png('Classification_Tree.png')
Image(graph.create_png())
```

# Random Forest
https://towardsdatascience.com/random-forest-in-python-24d0893d51c0

```python
# Initiatie RF with 10 decision trees (at each split, choose different predictors)
rf = RandomForestClassifier(n_estimators = 100, # Number of trees to grow
                            max_features = 'sqrt', # Number of different predictors chosen at each step
                            random_state = 232323, 
                            n_jobs = -1)
```

```python
# Train the model on Train Sample
rf.fit(x_train, y_train.values.ravel())
```

## Results

### Accuracy in Train

```python
# Make predictions in Training Set
yhat_train = rf.predict(x_train)
```

```python
metrics.accuracy_score(y_true = y_train, y_pred = yhat_train)
```

```python
# Compute Confusion Matrix for training set
cm = metrics.confusion_matrix(y_true = y_train, y_pred = yhat_train)
cm = pd.DataFrame(cm)
cm.columns = ["Predicted_0", "Predicted_1"]
cm.index = ["Actual_0", "Actual_1"]
cm
```

### Accuracy in Test

```python
# Make predictions in Testing Set
yhat_test = rf.predict(x_test)
```

```python
metrics.accuracy_score(y_true = y_test, y_pred = yhat_test)
```

```python
# Compute Confusion Matrix for testing set
cm = metrics.confusion_matrix(y_true = y_test, y_pred = yhat_test)
cm = pd.DataFrame(cm)
cm.columns = ["Predicted_0", "Predicted_1"]
cm.index = ["Actual_0", "Actual_1"]
cm
```

### Classification Report for Test Data

```python
print(metrics.classification_report(y_true = y_test, y_pred = yhat_test))
```

### Feature importance

```python
rf.feature_importances_
```

```python
features = feature_names
importances = rf.feature_importances_
indices = np.argsort(importances)

plt.title('Feature Importances')
plt.barh(range(len(indices)), importances[indices], color='b', align='center')
plt.yticks(range(len(indices)), [features[i] for i in indices])
plt.xlabel('Relative Importance')
plt.show()
```

# Bagging

```python
# Initialize Bagging Classifier
bag_clf = BaggingClassifier(base_estimator = tree_model, # Ensemble trees
                            n_estimators = 100, # Grow 10 trees
                            max_samples = 1.0, # Each tree, sample all rows (with replacement)
                            n_jobs= -1,
                            random_state = 232323)
```

```python
# Fit Bagging Classifier
bag_clf.fit(x_train, y_train.values.ravel())
```

## Results

```python
# Make predictions in Testing Set
yhat_test = bag_clf.predict(x_test)
```

```python
print(metrics.classification_report(y_true = y_test, y_pred = yhat_test))
```

# Boosting

```python
# Initialize Adaptative Boosting Classifier
ada_clf = AdaBoostClassifier(base_estimator = tree.DecisionTreeClassifier(max_depth=1), # Build one node at a time
                             n_estimators = 100, # Number of different trees to grow 
                             learning_rate = 0.5, # Reduce the contribution of each classifier, slower convergence
                             random_state = 232323) 
```

```python
# Fit adaboost
ada_clf.fit(x_train, y_train.values.ravel())
```

## Results

```python
# Make predictions in Testing Set
yhat_test = ada_clf.predict(x_test)
```

```python
print(metrics.classification_report(y_true = y_test, y_pred = yhat_test))
```
